```python
import numpy as np
import pandas as pd
from collections import Counter
import string
import re
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import plotly.express as px
```

# English

```python
tweets = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Copy of EnglishData.csv')
```

## Time Series Content Creation

```python
tweets['date'] = tweets.apply(lambda x: datetime.datetime.strptime(x['created_at'][:-17]+x['created_at'][-5:],'%a %b %d %H %Y')+datetime. timedelta(hours = 5, minutes = 30) ,axis=1)
frequencies = tweets['date'].value_counts()
d = frequencies.reset_index()

```

```python
# Save Specific Timing Tweets

tweets['in_peak'] = tweets.apply(lambda x:  datetime.time(7,30) <= x['date'].time() <=  datetime.time(8,30),axis=1)

filtered_tweets = tweets[tweets['in_peak']]

filtered_tweets.to_csv('/content/drive/MyDrive/Colab Notebooks/Filtered EnglishData.csv')
filtered_tweets['full_text'].to_csv('/content/drive/MyDrive/Colab Notebooks/Filtered EnglishData (text_only).csv')

# print(filtered_tweets.columns)
```

```python
# Timeseries Plots

time_ = []

for s_time in d['index']:
  time_.append(s_time)

d['Time'] = time_

d = d.rename(columns={'date':'tweets'})
d= d.sort_values(by=['Time'])


fig = px.line(d,x='Time',y='tweets')
fig.show()
```

## Time Series User Account Creation

```python
# User Account Creation Timeseries

user_accounts = tweets[['user.id_str','user.created_at']].drop_duplicates()
```

```python

user_accounts['date'] = user_accounts.apply(lambda x: datetime.datetime.strptime(x['user.created_at'][:-20]+x['user.created_at'][-5:],'%a %b %d %Y') ,axis=1)
frequencies = user_accounts['date'].value_counts()
d = frequencies.reset_index()
time_ = []

for s_time in d['index']:
  time_.append(s_time)
  
  time_[-1] = time_[-1].replace(minute=0,second=0,hour=0)

d['Time'] = time_
d= d.sort_values(by=['Time'])
# print(d)
d=d[1:]
fig = px.line(d,x='Time',y='date')
fig.show()
```
