```python
import textblob
from textblob import TextBlob
import seaborn as sns
import tqdm

def polarityscore(comment):
    return TextBlob(comment).sentiment.polarity

def subjectivityscore(comment):
    return TextBlob(comment).sentiment.subjectivity


dataset['Polarity']=dataset['full_text'].apply(polarityscore) 
dataset['Subjectivity']=dataset['full_text'].apply(subjectivityscore) 



dataset.head(30)
```

```python
print('Polarity Mean :')
print(dataset['Polarity'].mean(axis=0))
print('Subjectivity Mean  :')
print(dataset['Subjectivity'].mean(axis=0))
print('Polarity Standard Deviation :')
print(dataset['Polarity'].std(axis=0))
print('Subjectivity Standard Deviation :')
print(dataset['Subjectivity'].std(axis=0))
```
