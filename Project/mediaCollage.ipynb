```python
import pandas as pd
import numpy as np
import math
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from tqdm import tqdm
from collage_maker import make_collage
import csv 
import requests
from PIL import Image
```

```python
df = pd.read_csv('EnglishData.csv')
# df = pd.read_csv('HindiData.csv')
# df = pd.read_csv('PunjabiData.csv')
df.head()
```

```python
indexToDrop = []
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    #removing rows with null media entities and retweets
    media = row['extended_entities.media']
    if (type(media) == type(6.9) and math.isnan(row['extended_entities.media'])) or (row["full_text"][0:2] == "RT"):
        indexToDrop.append(index)
```

```python
#removing rows wihtout media
new_df = df.drop(df.index[indexToDrop])
```

```python
new_df
```

```python
#sorting rows on the basis of retweets
new_df = new_df.sort_values(by=['retweet_count'], ascending=False)
```

```python
#taking top 100 rows
new_df = new_df.head(100)
new_df
```

```python
mediaLinks = []
for index, row in new_df.iterrows():
    media = str(row['extended_entities.media'])
    #replacing the constants with python compatible constants
    media = media.replace("true", "True") 
    media = media.replace("false", "False") 
    media = media.replace("null", "None")
    #converting string to list containing dictionary (for media info)
    mediaDict = eval(media)
    mediaLinks.append(mediaDict[0]['media_url'])
```

```python
len(mediaLinks)
```

```python
mediaLinks
```

```python
#downloading images
listofimages=[]
count = 0
for url in mediaLinks:
    filename = "image"+str(count)+".jpg"
    count+=1
    result = requests.get(url, stream=True) 
    if result.status_code == 200: 
        image = result.raw.read()
        listofimages.append(filename)
        open(filename,"wb").write(image) 
```

```python
#creating collage
#reference: https://stackoverflow.com/a/35460517/10246960
def create_collage(width, height, listofimages):
    cols = 9
    rows = 9
    thumbnail_width = width//cols
    thumbnail_height = height//rows
    size = thumbnail_width, thumbnail_height
    new_im = Image.new('RGB', (width, height))
    ims = []
    for p in listofimages:
        im = Image.open(p)
        im.thumbnail(size)
        ims.append(im)
    i = 0
    x = 0
    y = 0
    for col in range(cols):
        for row in range(rows):
#             print(i, x, y)
            new_im.paste(ims[i], (x, y))
            i += 1
            y += thumbnail_height
        x += thumbnail_width
        y = 0

    new_im.save("Collage_punjabi.jpg")

create_collage(3840, 2160, listofimages)
```
