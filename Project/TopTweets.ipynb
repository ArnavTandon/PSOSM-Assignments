```python
def topvertweets(dataframe):
    dic = {}  
    for index, row in tqdm(dataframe.iterrows(), total=dataframe.shape[0]):
        x = row['full_text']
        y = row['retweet_count']
        if x in dic.keys():
            dic[x] += y
        else:
            dic[x] = y
        
    com = Counter(dic)
    temp = dict(com.most_common(10))
    for k in temp:
        print(k , temp[k])
        print("----------------------------------------==============---------------")
```

## Top engaged Tweets(English)

```python
from tqdm import tqdm
from collections import Counter
def toptweets(dataframe):
    full_text = dataframe['full_text']
    retweets = dataframe['retweet_count']
    
    dic = {}
    
    for i in tqdm(range(len(full_text))):
        if full_text[i] in dic.keys():
            dic[full_text[i]] += retweets[i]
        else:
            dic[full_text[i]] = retweets[i]
    
    x = Counter(dic)
    temp = dict(x.most_common(10))
    for k in temp:
        print(k , temp[k])
        print("----------------------------------------==============---------------")
    

toptweets(eng_df)
```

```python
import json
from collections import Counter
import codecs,string


def detect_language(character):
    maxchar = max(character)
    if u'\u0900' <= maxchar <= u'\u097f':
        return 'hindi'

    
def word(string):
    cloud = WordCloud(background_color = "black", height = 300, max_words = 10,stopwords = None)
    cloud.generate(string)

    # Display the generated image:
    plt.imshow(cloud)
    plt.axis("off")
    plt.show()

def extractTophashtags(hashtags):
    ex = []
    res = []
    for x in hashtags:
        l = []
        l.append(x)
        res = [json.loads(idx.replace("'", '"')) for idx in l]
        #print(res)
    
        for lx in res:
            for z in lx:
                for key in z:
                    if(key == 'text'):
                        if(detect_language(z[key])!= 'hindi'):
                            ex.append('#'+z[key])
    
    Count = Counter(ex)
    most_occur = dict(Count.most_common(10))
    return most_occur


def printdic(dic):
    txt = ''
    for x in dic:
        txt += x + " "
    return txt

hindi_hashtags = extractTophashtags(hindi_df['entities.hashtags'])

print('Top-Hindi-Hashtags')
print(" ------------------- ")


# word(printdic(hindi_hashtags))
# printdic(hindi_hashtags)
```
